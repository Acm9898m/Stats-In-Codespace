GEH	GEH	Classificação GEH
Problema: Nomenclaturas Incompatíveis
As colunas de volumes e sentidos usam nomenclaturas diferentes:

2020: A_C, B_C, ... J_C (Crescente) | A_D, B_D, ... J_D (Decrescente)

```r
# Compare o ano de 2020 e o de 2021 e me diga que colunas se repetem, 
# para que possamos juntar as duas tabelas num mesmo arquivo parquet particionado por ano.


################################################################################
#         UNIFICAR DATASETS 2020 E 2021 EM ARQUIVO PARQUET PARTICIONADO       #
################################################################################

# Preparar dataset 2020: renomear colunas de sentidos e adicionar metadados
df_trafego20_prep <- df_trafego20 %>%
  rename(
    # Renomear sentidos: Crescente → AB, Decrescente → BA
    A_AB = A_C, B_AB = B_C, C_AB = C_C, D_AB = D_C, E_AB = E_C,
    F_AB = F_C, G_AB = G_C, H_AB = H_C, I_AB = I_C, J_AB = J_C,
    A_BA = A_D, B_BA = B_D, C_BA = C_D, D_BA = D_D, E_BA = E_D,
    F_BA = F_D, G_BA = G_D, H_BA = H_D, I_BA = I_D, J_BA = J_D,
    VMDA_AB = VMDa_C,
    VMDA_BA = VMDa_D,
    UF = sg_uf,
    CLASSIFICACAO = classificação
  ) %>%
  # Selecionar apenas colunas comuns + metadados úteis
  select(
    ID = ID, CODIGO_BR = vl_br, REGIAO = NA_character_, UF, 
    EXTENSAO = vl_extensa, SUPERFICIE = ds_superfi,
    JURISDICAO = ds_jurisdi, TIPO_LINK = nm_tipo_tr,
    RELEVO_PRE = NA_character_, VELOCIDADE = NA_character_,
    CLASSIFICACAO, GEH,
    starts_with("A_AB"), starts_with("B_AB"), starts_with("C_AB"),
    starts_with("D_AB"), starts_with("E_AB"), starts_with("F_AB"),
    starts_with("G_AB"), starts_with("H_AB"), starts_with("I_AB"),
    starts_with("J_AB"), starts_with("A_BA"), starts_with("B_BA"),
    starts_with("C_BA"), starts_with("D_BA"), starts_with("E_BA"),
    starts_with("F_BA"), starts_with("G_BA"), starts_with("H_BA"),
    starts_with("I_BA"), starts_with("J_BA"),
    VMDA_AB, VMDA_BA
  ) %>%
  mutate(
    Ano = 2020,
    REGIAO = NA_character_,
    RELEVO_PRE = NA_character_,
    VELOCIDADE = NA_integer_
  )

# Preparar dataset 2021: adicionar coluna Ano
df_trafego21_prep <- df_trafego21 %>%
  mutate(
    Ano = 2021,
    # Calcular VMDA_TOTAL (2021 já possui, mas padronizamos)
    ID = ID_PNCT
  ) %>%
  select(
    ID, CODIGO_BR, REGIAO, UF, EXTENSAO, SUPERFICIE,
    JURISDICAO, TIPO_LINK, RELEVO_PRE, VELOCIDADE,
    CLASSIFICACAO, GEH,
    starts_with("A_AB"), starts_with("B_AB"), starts_with("C_AB"),
    starts_with("D_AB"), starts_with("E_AB"), starts_with("F_AB"),
    starts_with("G_AB"), starts_with("H_AB"), starts_with("I_AB"),
    starts_with("J_AB"), starts_with("A_BA"), starts_with("B_BA"),
    starts_with("C_BA"), starts_with("D_BA"), starts_with("E_BA"),
    starts_with("F_BA"), starts_with("G_BA"), starts_with("H_BA"),
    starts_with("I_BA"), starts_with("J_BA"),
    VMDA_AB, VMDA_BA, VMDA_TOTAL, Ano
  )

# Adicionar VMDA_TOTAL ao dataset 2020 (calcular como soma dos sentidos)
df_trafego20_prep <- df_trafego20_prep %>%
  mutate(
    VMDA_TOTAL = VMDA_AB + VMDA_BA,
    Ano = 2020
  )

# Unificar os dois datasets
df_trafego_unificado <- bind_rows(df_trafego20_prep, df_trafego21_prep)

cat("=== DATASET UNIFICADO ===\n")
cat("Dimensões:", nrow(df_trafego_unificado), "linhas x", 
    ncol(df_trafego_unificado), "colunas\n")
cat("Período: 2020 a 2021\n")
print(head(df_trafego_unificado, 3))

# Salvar em Parquet particionado por Ano
library(arrow)

write_dataset(
  df_trafego_unificado,
  path = "Trabalho/bases-de-dados/contagem-de-trafego/trafego_unificado_2020_2021",
  format = "parquet",
  partitioning = "Ano"
)

cat("\n✓ Dataset unificado salvo com particionamento por Ano\n")
```
